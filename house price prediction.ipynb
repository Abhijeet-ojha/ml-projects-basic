importing dependencies

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn.datasets
from sklearn.model_selection import train_test_split
from xgboost import XGBRegressor
from sklearn import metrics
```

```python
columns= ["CRIM","ZN","INDUS","CHAS","NOX","RM","AGE","DIS","RAD","TAX","PTRATIO","B" \
,"LSTAT","PRICE"]
house_data = pd.read_csv('C:/Users/Abhijeet ojha/Downloads/housing (2).csv', sep=r"\s+",          # split on any whitespace
    header=None,
    names=columns)
```

```python
print(house_data)
```

```python
#converting to pandas dataframe
house_data_df = pd.DataFrame(house_data)
```

```python
corr= house_data_df.corr()
plt.figure(figsize=(10,10))
sns.heatmap(corr, cbar=True, square=True, fmt='.2f',annot = True, annot_kws={'size':8},cmap='Blues')
```

splitting data 

```python
x = house_data_df.drop(['MEDV'], axis=1)
y = house_data_df['MEDV']
```

splitting into training and test

```python
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.2, random_state=2)
```

model training

```python
#loading model
model = XGBRegressor()
```

```python
#training using x_train
model.fit(x_train, y_train)
```

```python
training_data_prediction = model.predict(x_train)
print(training_data_prediction)
```

```python
# R squared error
score_1 = metrics.r2_score(y_train, training_data_prediction)
print(score_1)

# mean absolute error
score_2 = metrics.mean_absolute_error(y_train, training_data_prediction)
print(score_2)
```

actual vs predicted price


```python
plt.scatter(y_train, training_data_prediction)
plt.xlabel("Actual Prices")
plt.ylabel("Predicted Prices")
```

```python
#accuracy for prediction on test data
test_data_prediction = model.predict(x_test)
```

```python
# R squared error
score_1 = metrics.r2_score(y_test, test_data_prediction)
print(score_1)

# mean absolute error
score_2 = metrics.mean_absolute_error(y_test, test_data_prediction)
print(score_2)
```
